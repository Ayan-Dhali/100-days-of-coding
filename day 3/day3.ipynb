# Control Flow and Conditional 

<p style="font-size:20px">Suppose you are playing a game and you have two choice either go left or right</p>


##### ***You use if-else statement in this case***
Syntax
``` 
if condition:
    this
else:
    this
```

````python
age=int(input("Enter your age?"))
if age>=18:
    print("Adult")
else:
    print("child")
````

##### ***Now suppose you have more than two condition,like if age<13 child if age>13 but age<18 teen and if age>18 adult***
Now in this case we use ***nested loop***
Syntax
```
if condition:
     this
else if condition:
    this
...
else:
    this
```

````python
age=int(input("Enter your age?"))
if age>=18:
    print("Adult")
elif age>=13 and age<18:
    print("Teen")
else:
    print("child")
````
